###  Identify gender subspace.

```python
 from sklearn.decomposition import TruncatedSVD
sum_w = 0.0
for job in occupation_list:
    sum_w+=norm_wv[job]
mean_w = sum_w/float(len(occupation_list))
C = 0.0
for job in occupation_list:
    w_minus_u = np.expand_dims((norm_wv[job]-mean_w), axis=0)
    C+= (w_minus_u.T*w_minus_u)/float(len(occupation_list))
print(C.shape)
svd = TruncatedSVD(n_components=3, n_iter=7, random_state=42)  
svd.fit(C)
```

```python
svd.components_[0].shape
```

```python
w_b = {}
for job in occupation_list:
    w = norm_wv[job]
    w_b[job]=0.0
    for component in svd.components_:
        w_b[job] += np.dot(norm_wv[job],component)*component
```
